### 3.5 Categorical Plots

```python
# Count the number of distinct tsunami events per country
country_counts = runups.groupby('country')['tsunami_event_id'].nunique().reset_index()

# Filter countries with more than 200 tsunami events
country_counts_filtered = country_counts[country_counts['tsunami_event_id'] > 40]

# Basic Bar Chart with count of tsunami events per country (for countries with more than 10 events)
plt.figure(figsize=(10, 6))
sns.barplot(x='country', y='tsunami_event_id', data=country_counts_filtered)
plt.title('Tsunami Events by Country (More than 40 Events)')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

**Observation:** The top countries for tsunami events in the last 40 years are Chile, Japan, New Zealand, USA, USA Territories (be interesting to see more detail on this) and Vanuatu.

Interestingly, Indonesia, the location of the boxing day tsunami that began this analysis, is not featured in this list.

```python
# Count the number of unique tsunami events per country
country_counts = runups.groupby('country')['tsunami_event_id'].nunique().reset_index()

# Filter countries with more than 40 tsunami events
country_counts_filtered = country_counts[country_counts['tsunami_event_id'] > 40]

# Count the occurrences of each earthquake severity per country, using unique tsunami events
# Add observed=False to prevent the FutureWarning
severity_counts = runups.groupby(['country', 'flag_earthquake_severity'], observed=False)['tsunami_event_id'].nunique().reset_index(name='severity_count')

# Merge the filtered country counts with the severity counts
merged_data = pd.merge(severity_counts, country_counts_filtered, on='country')

# Create a pivot table to reshape the data for the stacked bar chart
pivot_data = merged_data.pivot_table(index='country', columns='flag_earthquake_severity', values='severity_count', aggfunc='sum', fill_value=0, observed=False)

# Plot the stacked bar chart
ax = pivot_data.plot(kind='bar', stacked=True, figsize=(14, 7))

# Title and labels
plt.title('Earthquake Severity Distribution by Country (More than 40 Tsunami Events)', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Tsunami Events', fontsize=12)

# Rotate x-axis labels
plt.xticks(rotation=90)

# Move legend outside the plot
plt.legend(title='Earthquake Severity', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add percentages inside the bars
for i, p in enumerate(ax.patches):
    # Find the total height for each country (sum of the stacked bars)
    total_height = sum([patch.get_height() for patch in ax.patches if patch.get_x() == p.get_x()])
    
    # Height and width of the current segment
    height = p.get_height()
    
    # Calculate the percentage for the current segment
    percentage = (height / total_height) * 100 if total_height != 0 else 0
    
    # Calculate position for the text
    x = p.get_x() + p.get_width() / 2  # X position for the text (center of the bar)
    y = p.get_y() + p.get_height() / 2  # Y position for the text (center of the bar)

    # Display the percentage inside the bar
    ax.text(x, y, f'{percentage:.1f}%', ha='center', va='center', color='white', fontsize=10)

# Show the plot
plt.show()
```

**Observations:** We can see that all of these countries experience earthquakes from at least 6.1 or higher. The most common earthquake magnitude for these countries is in the range of 7- 7.9.
